```python
import matplotlib.pyplot as plt

y=[95, 90, 85, 88, 98]
x=["Part1", "Part2", "Part3", "Part4", "Part5"]
plt.bar(x, y, color= "blue", edgecolor="black", width= 0.8)
plt.xlabel("Parts of movie", fontsize= 12)
plt.ylabel("Popularity",fontsize= 12 )
plt.title("POPULARITY OF DIFFERENT PARTS OF MOVIE", fontsize= 17)
plt.show() 
```

```python
import pandas as pd

data= pd.read_csv("expenses.csv")
df= pd.DataFrame(data)
# print(df)
plt.bar(df["Payment Mode"], df["Amount"])
plt.show()
```

```python
# using group_by

import pandas as pd

data= pd.read_csv("expenses.csv")
df= pd.DataFrame(data)
# print(df)

gp = df.groupby("Payment Mode")["Amount"].sum()
print(gp)

plt.bar(gp.index, gp.values)
plt.title("Amount with different payment mode ", color= "green")
plt.show()
```

```python
x= ["day1", "day2", "day3", "day4", "day5"]
y=[300,410,500,380,400]
y1= [360,290,450,400,430]

# plt.plot(x, y, marker="o", ls= ":", color= "red", label="Week 1")

# to show transparency in week 1
plt.plot(x, y, marker="o", ls= ":", color= "red", label="Week 1", alpha= 0.5)

plt.plot(x, y1, marker="*", ls= "-", color= "green", label="Week 2")
plt.title("LINE CHART")
plt.legend()
plt.show()
```
