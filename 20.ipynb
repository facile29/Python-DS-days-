```python
import matplotlib.pyplot as plt 

x= [1, 2, 3, 4, 5, 6]
y= [45, 35, 66, 55, 34, 60]
plt.subplot(1, 2, 1)
plt.bar(x, y)
plt.title("age")

x= [1, 2, 3, 4, 5, 6]
y= [67, 88, 43, 48, 58, 60]
plt.subplot(1, 2, 2)
plt.title("weight")
plt.plot(x, y) 
plt.suptitle("SUBPLOTS")
plt.show()
```

```python
import matplotlib.pyplot as plt 

x= [1, 2, 3, 4, 5, 6]
y= [45, 35, 66, 55, 34, 60]
plt.subplot(2, 2, 1)  
plt.bar(x, y)

x= [5, 6, 7, 8, 9, 10]
y= [67, 88, 43, 48, 58, 60]
plt.subplot(2, 2, 2)
plt.plot(x, y) 

x= [2, 4, 6, 8, 10, 12]
y= [37, 38, 47, 50, 48, 66]
plt.subplot(2, 2, 3)
plt.scatter(x, y) 

x= [1, 3, 5, 7, 9, 11]
y= [37, 38, 47, 50, 48, 66]
plt.subplot(2, 2, 4)
plt.step(x, y) 

plt.suptitle("Different Charts")
plt.show()
```
