```python
import numpy as np 
import matplotlib.pyplot as plt

x= np.random.randint(1, 10, 50)
y= np.random.randint(10, 100, 50)
color=np.random.randint(10, 100, 50)

plt.scatter(x, y, marker="*", cmap= "hot", c= color)
plt.colorbar()
plt.title("SCATTER PLOT")
plt.show()
```

```python
brands=["apple", "oneplus", "redmi", "nokia", "samsung"]
x=[35, 30, 22, 3, 10]
plt.pie(x)
plt.title("PIE CHART")
plt.show()
```

```python
brands=["apple", "oneplus", "redmi", "nokia", "samsung"]
x=[35, 30, 22, 3, 10]
ex= [0, 0.2, 0, 0, 0]
plt.pie(x, labels= brands, explode= ex, shadow= True, autopct="%.2f", startangle= 180)
plt.show() 
```

```python
data= pd.read_csv("expenses.csv")
df= pd.DataFrame(data)
gp = df.groupby("Payment Mode") ["Amount"].sum()
print(gp)
plt.pie(gp.values, labels= gp.index, autopct= "%.2f")
plt.title("PIE CHART", color="red", size= 15)
plt.show()
```
