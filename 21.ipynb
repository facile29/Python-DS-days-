```python
# lineplot in seaborn

import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt

data={"days":[1, 2,3, 4, 5],
      "NOP":[38, 28, 22, 40, 25]}
df= pd.DataFrame(data)
# print(df)

sns.lineplot(data= data, x="days", y="NOP")
plt.show()
```

```python
data= pd.read_csv("expenses.csv")
# print(data)
sns.lineplot(data=data, x="Category", y="Amount")
plt.show()
```

```python
data= pd.read_csv("expenses.csv")
# print(data)
sns.lineplot(data=data, x="Category", y="Amount", hue="Payment Mode")
plt.show()
```

```python
data= pd.read_csv("expenses.csv")
# print(data)
sns.lineplot(data=data, x="Category", y="Amount", style= "Payment Mode")
plt.show()
```

```python
data= pd.read_csv("expenses.csv")
# print(data)
c= sns.color_palette("hsv", 3)
sns.lineplot(data=data, x="Category", y="Amount", hue= "Payment Mode",palette=c)
plt.show()
```

```python
# bar plot in seaborn 

data = sns.load_dataset("tips")
print(data)
sns.barplot(data= data, x="day", y="tip")

plt.plot()
```

```python
# bar plot in seaborn 

data = sns.load_dataset("tips")
print(data)
sns.barplot(data= data, x="day", y="tip", estimator= "mean", hue="sex", palette="viridis")

plt.plot()
```

```python
# bar plot in seaborn 

data = sns.load_dataset("tips")
print(data)
sns.barplot(data= data, x="day", y="tip", estimator= "mean", hue="sex", palette="spring", errorbar=("ci", 0))

plt.plot()
```

```python
# Histogram plot 

df= sns.load_dataset("titanic")
# print(df)
sns.histplot(data= df, x="age", hue="sex", kde= True, palette= "gist_earth")
plt.plot()
```

```python
# Histogram plot 

df= sns.load_dataset("titanic")
# print(df)
sns.histplot(data= df, x="age", hue="who", kde= True,bins= 30, palette= "turbo")
plt.plot()
```
