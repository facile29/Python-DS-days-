```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset("tips")
print(data)
```

```python
sns.scatterplot(data= data, x="total_bill", y="tip")
plt.show()
```

```python
sns.scatterplot(data= data, x="total_bill", y="tip", hue="smoker", size= "size", marker="D", palette="viridis")
plt.legend(bbox_to_anchor = (0.8, 0., 0.5, 1.0))
plt.show()
```

```python
data= pd.read_excel("employee.xlsx")
#print(data)

sns. scatterplot(data= data, x="Age", y="Annual Salary", size="Bonus %", hue="Department")
plt.legend(bbox_to_anchor = (0.2, 0., 1.2, 1))
plt.show()
```

```python
data= sns.load_dataset("tips")
# print(data)
gp= data.groupby("day").agg({"tip":"mean"})
sns.heatmap(data= gp, annot=True)
plt.show()
```

```python
data= pd.read_excel("employee.xlsx")
#print(data)
gp= data.groupby("Job Title").agg({"Annual Salary":"mean"})
sns.heatmap(data= gp)
plt.show()
```

```python
data= sns.load_dataset("tips")
sns.countplot(data= data, x="day")
plt.show()
```

```python
data= sns.load_dataset("tips")
sns.countplot(data= data, x="sex",hue="sex" ,palette=["blue", "orange"])
plt.show()
```

```python
data= pd.read_excel("employee.xlsx")
sns.countplot(data=data, x="Ethnicity", hue="Gender", palette ="viridis")
plt.show()
```
