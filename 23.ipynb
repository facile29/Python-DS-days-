```python
import matplotlib.pyplot as plt
import seaborn as sns

data= sns.load_dataset("tips")
sns.catplot(data= data, x="day", y="total_bill", hue="sex", kind= "violin")
plt.show()
```

```python
data= sns.load_dataset("tips")
sns.catplot(data= data, x="day", y="total_bill", hue="sex", kind= "bar")
plt.show()
```

```python
# background color
data= sns.load_dataset("tips")
sns.set_style(style= "whitegrid")
sns.catplot(data= data, x="day", y="total_bill", hue="sex", kind= "bar", palette="viridis")
plt.show()
```

```python
# background color
data= sns.load_dataset("tips")
sns.set_style(style= "ticks")
sns.catplot(data= data, x="day", y="total_bill", hue="sex", kind= "bar", palette="spring")
plt.show()
```

```python
# multiple plot in seaborn 

df= sns.load_dataset("tips")
# print(df)
order = ['Thur', 'Fri', 'Sat', 'Sun']
a= sns.FacetGrid(data= df, col="smoker", height=3)
a.map(sns.barplot, "day", "tip", order=order)
plt.show()
```

```python
# multiple plot in seaborn 

df= sns.load_dataset("tips")
# print(df)
order = ['Thur', 'Fri', 'Sat', 'Sun']
a= sns.FacetGrid(data= df, col="time", height=3, hue="sex")
a.map(sns.barplot, "day", "tip", order= order)
plt.show()
```

```python
# relational plot
df= sns.load_dataset("tips")
sns.relplot(data= data, x="tip", y="total_bill", hue="sex", kind="line", col="day")
plt.show()
```

```python
# relational plot
df= sns.load_dataset("tips")
sns.relplot(data= data, x="tip", y="total_bill", hue="sex")
plt.show()
```

```python
# Swarm Plot

df= sns.load_dataset("tips")
sns.swarmplot(data= df, x="day", y="total_bill", hue="sex", dodge= True)
plt.legend(bbox_to_anchor= (0.8, 0, 0.5, 1.0))
plt.show()
```

```python
# KDEPlot 

df= sns.load_dataset("tips")
sns.kdeplot(data=df, x="total_bill", hue="day")
plt.show()
```

```python
# KDEPlot 

df= sns.load_dataset("tips")
sns.kdeplot(data=df, x="total_bill", hue="day", multiple="stack")
plt.show()
```
