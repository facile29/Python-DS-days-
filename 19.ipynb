```python
import pandas as pd
import matplotlib.pyplot as plt

x= [10, 40,80, 60,20, 50, 40, 0, 30, 70, 10]
plt.stem(x)
plt.title("Stem Plot")
plt.show()
```

```python
x= [10, 40,80, 60,20, 50, 40, 0, 30, 70, 10]
plt.stem(x, linefmt="--", markerfmt="D", orientation ="horizontal")
plt.title("Stem Plot in Horizontal")
plt.show()
```

```python
x= [10, 40,80, 60,20, 50, 40, 0, 30, 70, 10]
plt.stem(x, linefmt=":", markerfmt="^", orientation ="vertical")
plt.title("Stem Plot in Vertical")
plt.show()
```

```python
x= [10, 40,80, 60,20, 50, 40, 0, 30, 70, 10]
plt.plot(x)
plt.stem(x, linefmt="--", markerfmt="D", orientation ="vertical")
plt.title("Stem Plot with Line Plot")
plt.show()
```

```python
days= [1, 2, 3, 4, 5, 6, 7]
n1= [10, 5, 70, 30, 35, 80, 50]
n2= [20, 40, 30, 55, 12, 8, 30]
n3= [8, 10, 35, 55, 65, 80, 75]
# c= ["pink", "hot pink", "magenta"]
plt.stackplot(days, n1, n2, n3, colors = ["pink", "hotpink", "magenta"], labels=["week 1", "week 2", "week 3"])
plt.legend()
plt.title("Satck Plot")
plt.show()
```

```python
# step plot
x= ["day1", "day2", "day3", "day4", "day5"]
y= [30, 40, 25, 35, 40]
plt.step(x, y)
plt.show()
```

```python
# legend matplotlib
x= [1, 2, 3, 4, 5]
y1= [45, 67, 33, 70, 12]
y2= [42, 76, 33, 64, 13]
y3= [40, 62, 35, 80, 15]
plt.figure(figsize= [5, 3])
plt.plot(x, y1, label="male")
plt.plot(x, y2, label="female")
plt.plot(x, y3, label="other")
plt.legend( bbox_to_anchor= (1.4, 1.2))
plt.show()
```
